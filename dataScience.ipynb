```python
# program 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("iris.csv")
plt.figure(figsize=(10,5))
x=data["sepal.length"]
plt.hist(x,bins=20,color="blue")
plt.title("Sepal length in cm")
plt.xlabel("Sepal length in cm")
plt.ylabel("count")
plt.show()
```
